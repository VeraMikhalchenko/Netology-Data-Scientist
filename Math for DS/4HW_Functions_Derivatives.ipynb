Посчитайте значение первой и второй производной функции:

f(x)=x**5+4sin(2x)+cos(3x+3)

В точке x = 1.

```python
import numpy as np
import sympy as sym
import math
import matplotlib.pyplot as plt
```

```python
x = sym.Symbol('x')
y = x**5 + 4 * sym.sin(2*x) + sym.cos(3*x + 3)
yprime = sym.diff(y)
yprime
```

```python
x = 1
print('Значение первой производной в x=1:', round(5*(x**4) - 3*math.sin(3*x + 3) + 8*math.cos(2*x), 2))
    
```

```python
ysecond = sym.diff(yprime)
ysecond
```

```python
x = 1
print('Значение второй производной в x=1:',round(20*(x**3) - 16*math.sin(2*x) - 9*math.cos(3*x + 3), 2))
```

Постройте график первой и второй производной функции:

f(x)=sin**5(2x+1)

На отрезке [-5; 5].


```python
from pylab import rcParams
rcParams['figure.figsize'] = 12,7
```

```python
x = sym.Symbol('x')
y = sym.sin(2*x+1)**5
yprime = sym.diff(y)
ysecond = sym.diff(yprime)
print(f'первая производная: {yprime}, вторая производная: {ysecond}')
```

```python
y1 = []
y2 = []
x = np.arange(-5, 5, 0.01)

for i in x:
    yprime = 10*math.sin(2*i + 1)**4*math.cos(2*i + 1)
    y1.append(yprime)
    ysecond = -20*math.sin(2*i + 1)**5 + 80*math.sin(2*i + 1)**3*math.cos(2*i + 1)**2
    y2.append(ysecond)

plt.plot(x, y1)
plt.plot(x, y2)
plt.xlim([-5, 5])
plt.legend(['Первая производная', 'Вторая производная'], fontsize = 14)
```
