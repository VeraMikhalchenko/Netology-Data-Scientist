**Задание 1**\
В магазин привезли устройства с 3-х разных предприятий.

Соотношение устройств следующее: 
- 20% - продукция первого предприятия, 
- 30% - продукция второго предприятия, 
- 50% - продукция третьего предприятия; 
- далее, 10% продукции первого предприятия высшего сорта, на втором предприятии - 5% и на третьем - 20% продукции высшего сорта.

Найти вероятность того, что случайно купленная новая продукция окажется высшего сорта.

```python
P1 = 0.2 #вероятность покупки продукции с первого предприятия
P2 = 0.3 #вероятность покупки продукции со второго предприятия
P3 = 0.5 #вероятность покупки продукции с третьего предприятия

P1_h = P1 * 0.1 # вероятность покупки продукции высшего сорта с первого предприятия
P2_h = P2 * 0.05 # вероятность покупки продукции высшего сорта со второго предприятия
P3_h = P3 * 0.2 # вероятность покупки продукции высшего сорта с третьего предприятия

# Вероятность того, что случайно купленная новая продукция окажется высшего сорта, является суммой вероятностей:
print(P1_h + P2_h + P3_h)
```

Изучите 3 распределения из библиотеки scipy: Нормальное, Экспоненциальное, Стьюдента. Используя функцию плотности распредления (pdf), постройте пять графиков плотностей для каждого распределения при разных параметрах. Запишите ваши наблюдения: как и на что влияют параметры у каждого распределения.

Дополнительная информация тут.

```python
import numpy as np
from scipy.stats import norm
import matplotlib.pyplot as plt
```

**Нормальное распределение:**

```python
x = np.linspace(norm.ppf(0.01), norm.ppf(0.99), 100)
plt.plot(x, norm.pdf(x, loc = 0, scale = 0.5),'r-', label='norm 1')
plt.plot(x, norm.pdf(x, loc = 1, scale = 0.5),'k-', label='norm 2')
plt.plot(x, norm.pdf(x, loc = 0, scale = 1),'b-', label='norm 3')
plt.plot(x, norm.pdf(x, loc = -0.5, scale = 0.7),'y-', label='norm 4')
plt.plot(x, norm.pdf(x, loc = 0, scale = 0.3),'g', label='norm 5')

plt.legend(frameon=False)
```

Параметр loc указывает среднее значение, а параметр scale - стандартное отклонение. Чем меньше стандартное отклонение, тем более вытянутым по оси y будет график.

**Экспоненциальное распределение:**

```python
from scipy.stats import expon
```

```python
x = np.linspace(expon.ppf(0.01), expon.ppf(0.99), 100)
lambda_ = 1
plt.plot(x, expon.pdf(x, loc = 0, scale = 1/lambda_),'r-', label='expon 1')
lambda_ = 0.2
plt.plot(x, expon.pdf(x, loc = 0, scale = 1/lambda_),'k-', label='expon 2')
lambda_ = 3
plt.plot(x, expon.pdf(x, loc = 0, scale = 1/lambda_),'b-', label='expon 3')
lambda_ = 2
plt.plot(x, expon.pdf(x, loc = 0, scale = 1/lambda_),'y-', label='expon 4')
lambda_ = 0.5
plt.plot(x, expon.pdf(x, loc = 0, scale = 1/lambda_),'g-', label='expon 5')

plt.legend(frameon=False)
```

y = lambda при х = 0, при этом стандартное отклонение уменьшается при увеличении lambda, что собственно следует из формулы..

**Распределение Стьюдента:**

```python
from scipy.stats import t
```

```python
x = np.linspace(t.ppf(0.01, df = 5), t.ppf(0.99, df = 5), 100)
```

```python
plt.plot(x, t.pdf(x, df = 3),'r-', label='t 1')
plt.plot(x, t.pdf(x, df = 0.5, loc = -1, scale = 2),'k-', label='t 2')
plt.plot(x, t.pdf(x, df = 100),'b-', label='t 3')
plt.plot(x, t.pdf(x, df = 1),'y-', label='t 4')
plt.plot(x, t.pdf(x, df = 0.01),'g-', label='t 5')

plt.legend(frameon=False)
```

При показателе степеней свободы стремящемся к бесконечности распределение переходит в нормальное:

```python
x = np.linspace(norm.ppf(0.01), norm.ppf(0.99), 100)
plt.plot(x, norm.pdf(x, loc = 0, scale = 1),'r-', label='norm')

df = 10
x = np.linspace(t.ppf(0.01, df), t.ppf(0.99, df), 100)
plt.plot(x, t.pdf(x, df),'g-', label='t1')

df = 100
x = np.linspace(t.ppf(0.01, df), t.ppf(0.99, df), 100)
plt.plot(x, t.pdf(x, df),'b-', label='t2')


plt.legend(frameon=False)
```
