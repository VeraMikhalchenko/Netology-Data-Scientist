```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as sts
%matplotlib inline
```

```python
rv = sts.t(loc=1, scale = 1, df = 15)
sample = rv.rvs(size=3000)
```

```python
x = np.linspace(0,2,100)
pdf = rv.pdf(x)
plt.hist(sample, density=True, bins='auto', label='sample histogram')
plt.plot(x, pdf, label='theoretical pdf', alpha=0.5)
plt.legend()
plt.ylabel('fraction of samples, f(x)')
plt.xlabel('x')
plt.show()
```

```python
def normal_dist(x):
    """ Функция проверяет распредление данных на нормальность
    с помощью теста Шапиро-Уилка
    """
    result = []
    for data in x:       
        stat, p = sts.shapiro(data)

        print('stat=%.3f, p=%.3f' % (stat, p))
        if p > 0.05:
            result.append('Нормальное')
        else:
            result.append('Ненормальное')
    
    return result
```

```python
normal_dist([sample])
```

```python
n1 = 300            # Размеры выборок 
a1 = np.array([])     # Массивы для сохранения средних по выборкам
samples_count = 1000  # Количество повторов генерации

for number in range(samples_count): 
    sample1 = rv.rvs(size=n1)               # На каждом цикле генерируем выборки указанных размеров  
    a1 = np.append(a1, sample1.mean())      # Считаем среднее по каждой выборке и добавляем в массив средних
  
```

```python
label_1 = 'sample of ' + str(n1) + ' histogram'   # Подписи для графиков
plt.hist(a1, density=True, label=label_1)         # Строим графики

plt.legend() 
plt.ylabel('samples means')
plt.xlabel('x')
```

```python
normal_dist([a1])
```
