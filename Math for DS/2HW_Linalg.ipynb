Придумайте матрицу, у которой определитель = 1, и все элементы не равны 0.

P.S. Матрица должна быть ненулевой и размерностью не менее 2!

m = np.array([[a,b], [c,d]])\
det = a * d - c * b

В нашем случае определитель = 1, тогда

a * d = 1 + c * b

Можем взять два любых числа для координат векторов и на основе их вычислить остальные.
Предположим, что a = 2, d = 3, тогда

2 * 3 = 1 + c * b\
1 + c * b = 6\
c * b = 5

Тогда, например, подходит матрица:
m = np.array([[2,1],[5,3]])

Проверим
