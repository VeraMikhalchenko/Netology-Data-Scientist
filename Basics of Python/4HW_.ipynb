Домашнее задание к лекции "Функции"

Вам нужно помочь секретарю автоматизировать работу. Для этого нужно написать программу, которая будет на основе хранимых данных исполнять пользовательские команды.

Исходные данные имеют следующую структуру:

перечень всех документов
documents = [
 {'type': 'passport', 'number': '2207 876234', 'name': 'Василий Гупкин'},
 {'type': 'invoice', 'number': '11-2', 'name': 'Геннадий Покемонов'},
 {'type': 'insurance', 'number': '10006', 'name': 'Аристарх Павлов'}
]
перечень полок, на которых хранятся документы (если документ есть в documents, то он обязательно должен быть и в directories)
directories = {
 '1': ['2207 876234', '11-2'],
 '2': ['10006'],
 '3': []
}
Общие требования к программе:

код должен быть грамотно декомпозирован (каждая функция отвечает за свою конкретную задачу, дублирующийся функционал переиспользуется, а его код не повторяется);
в коде отсутствуют глобальные переменные (за исключением documents и directories);
пользовательский ввод обрабатывается в цикле while до тех пор, пока пользователь явно не завершит программу (вводом команды "q").

Исходные данные:

```python
documents = [
 {'type': 'passport', 'number': '2207 876234', 'name': 'Василий Гупкин'},
 {'type': 'invoice', 'number': '11-2', 'name': 'Геннадий Покемонов'},
 {'type': 'insurance', 'number': '10006', 'name': 'Аристарх Павлов'}
]

directories = {
 '1': ['2207 876234', '11-2'],
 '2': ['10006'],
 '3': []
}
```

Задание 1

Пункт 1. Пользователь по команде "p" может узнать владельца документа по его номеру

```python
def input_check(el, data):
        """
        Функция проверяет наличие документа в данных
        """
        for k in data.values():
                if el in k:
                        return True
        else:
                return False
```

```python
def people():
        """ 
        Функция p выводит на экран имя владельца документа, 
        номер которого введён пользователем,
        или сообщает о том, что документ не найден.
        """
        number = (input('Введите номер документа: '))

        if input_check(number, directories) == True:
                for num in documents:
                        if num['number'] == number:
                                print('Владелец документа:', num['name'])
                                return main()
                                 
        else:
                print('Документ не найден в базе')
                return main()
```

Пункт 2. Пользователь по команде "s" может по номеру документа узнать на какой полке он хранится

```python
def input_shelf_check(el,data):
        """
        Функция проверяет наличие полки в данных
        """
        for k in data.keys():
                if el in k:
                        return True
        else:
                return False
                
```

```python
def shelf():
        """ 
        Функция s выводит на экран номер полки документа, 
        номер которого введён пользователем,
        или сообщает о том, что документ не найден.
        """
        number = (input('Введите номер документа: '))

        if input_check(number, directories) == True:
                for key, value in directories.items():
                        if number in value:
                                print('Номер полки:', key)
                                return main()
      
        else:
                print('Документ не найден в базе')
                return main()
```

Пункт 3. Пользователь по команде "l" может увидеть полную информацию по всем документам

```python
def list_(docs, dirs):
        """ 
        Функция l выводит на экран полную информацию по всем документам
        """
        for num in docs:
              number_doc = num['number']
              type_ = num['type']
              owner = num['name']
              for k,v in dirs.items():
                      if number_doc in v:
                              print(f'№: {number_doc}, тип: {type_}, владелец: {owner}, полка хранения: {k}')
        return main()
```

```python
def add_shelf():
        """
        Функция добавляет новую полку
        """
        shelf = input('Введите номер полки: ')
        shelfs = [i for i in directories.keys()]

        if input_shelf_check(shelf,directories) == True:
                print('Такая полка уже существует. Текущий перечень полок: ',', '.join(shelfs))
                return main()

        else:
                directories.setdefault (shelf, [])
                shelfs = [i for i in directories.keys()]
                print('Полка добавлена. Текущий перечень полок: ',', '.join(shelfs))
                return main()
```

```python
def delete_shelf():
        """
        Функция удаляет существующую полку по номеру.
        """
        shelf = input('Введите номер полки: ')
        shelfs = [i for i in directories.keys()]

        if input_shelf_check(shelf, directories) == False:
                print('Нет такой полки. Текущий перечень полок: ', ', '.join(shelfs) )
                return main()
        for key, value in directories.items():
                if directories[shelf] == []:
                        del directories[shelf]
                        shelfs = [i for i in directories.keys()]
                        print('Полка удалена. Текущий перечень полок: ',', '.join(shelfs))
                        return main()
                else:
                        print('На полке есть документы, удалите их перед удалением полки. Текущий перечень полок: ',\
                              ', '.join(shelfs))
                        return main()
```

Дополнительные задания:

Пункт 1. Пользователь по команде "ad" может добавить новый документ в данные

```python
def add_doc():
        """
        Функция добавляет документ на полку
        """
        doc = input('Введите номер документа: ')
        
        if input_check(doc, directories) == True:
                print('Этот номер уже в базе')
                return main()

        type_doc = input('Введите тип документа: ')
        name = input('Введите владельца документа: ')
        shelf = input('Введите номер полки для документа: ')
        
        new_doc = {}

        new_doc['type'] = type_doc
        new_doc['number'] = doc
        new_doc['name'] = name
        
        documents.append(new_doc)
        
        for k,v in directories.items():
                if k == shelf:
                        v.append(doc)
        
        return list_(documents, directories)
        
```

Пункт 2. Пользователь по команде "d" может удалить документ из данных

```python
def del_doc():
        """
        Функция удаляет документ из данных
        """
        doc = input('Введите номер документа: ')
        
        if input_check(doc, directories) == False:
                print('Этого номера нет в базе.')
                print('Текущий список документов:')
                return list_(documents, directories)
        else:
                for k in directories.values():
                        for s in k:
                                if s == doc:         
                                        k.remove(s)
                for d in documents:
                        if doc in d.values():
                                documents.remove(d)
                        
                print('Документ удален.')
                print('Текущий список документов:')
                return list_(documents, directories)
```

Пункт 3. Пользователь по команде "m" может переместить документ с полки на полку

```python
def move_doc():
      doc = input('Введите номер документа: ')
      shelfs = [i for i in directories.keys()]

      if input_check(doc,directories) == False:
              print('Этого номера нет в базе.')
              print('Текущий список документов:')
              return list_(documents, directories)
      else:
              shelf = input('Введите номер полки, куда переместить документ: ')
              
              if input_shelf_check(shelf, directories) == False:
                      print('Этой полки нет в базе.')
                      print('Текущий перечень полок:', ', '.join(shelfs))
                      return main()
              
              else:
                      for k,v in directories.items():
                              if k == shelf and doc in v:
                                      print('Документ уже на этой полке.')
                                      print('Текущий список документов:')
                                      return list_(documents, directories)
                                              
                              elif k != shelf and doc in v:
                                      directories[shelf].append(doc)
                                      v.remove(doc)                                      
                                      print(f'Документ перемещен на полку {shelf}.')
                                      print('Текущий список документов:')
                                      return list_(documents, directories)
                              else:
                                      return 'impossible'
```

Главная функция:

```python
def main():
        """
        Главная функция. Пользователю предлагается ввести одну из команд для работы с данными.
        p - вызывает функцию people, которая по номеру документа выводит имя владельца
        s - вызывает функцию shelf, которая по номеру документа выводит номер его полки
        l - вызывает функцию list_ с двумя входящими переменными, которая выводит всю информацию по документам
        ads - вызывает функцию add_shelf, которая добавляет новую полку
        ds - вызывает функцию delete_shelf, которая удаляет пустую полку
        ad - вызывает функцию add_doc, которая добавляет новый документ
        d - вызывает функцию delete_doc, которая удаляет документ
        m - вызывает функцию move_doc, которая перемещает документ с полки на полку
        q - выход из программы
        """
        command = input('Введите команду (p, s, l, ads, ds, ad, d, m, q): ')

        while command != 'q':
                if command == 'p':
                        return people()    
                elif command == 's':
                        return shelf()       
                elif command == 'l':
                        return list_(documents, directories)
                elif command == 'ads':
                        return add_shelf()
                elif command == 'ds':
                        return delete_shelf()
                elif command == 'ad':
                        return add_doc()
                elif command == 'd':
                        return del_doc()
                elif command == 'm':
                        return move_doc()   
                else:
                        print('Неверный ввод')
                        return main()
        else:
                print('Программа завершена')
                return 

main()
```
