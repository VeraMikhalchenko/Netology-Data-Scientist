Вам необходимо написать функцию, которая будет основана на поиске по сайту habr.com. Функция в качестве параметра должна принимать список запросов для поиска (например, ['python', 'анализ данных']) и на основе материалов, попавших в результаты поиска по каждому запросу, возвращать датафрейм вида:

<дата> - <заголовок> - <ссылка на материал> \
В рамках задания предполагается работа только с одной (первой) страницей результатов поисковой выдачи для каждого запроса. Материалы в датафрейме не должны дублироваться, если они попадали в результаты поиска для нескольких запросов из списка.

```python
import pandas as pd
import requests
import time
from bs4 import BeautifulSoup
import urllib.parse
```

```python
res = requests.get('https://habr.com/ru/all/')
res.status_code
```

```python
url = 'https://habr.com/ru/all/'
params = {
    'q': 'python'
}

req = requests.get(url, params=params)
time.sleep(0.3)
soup = BeautifulSoup(req.text)
articles = soup.find_all('article', class_='tm-articles-list__item')
len(articles)
```

```python
def get_habr_posts(query):
    habr_article = pd.DataFrame()
  
    for x in query:
        x = urllib.parse.quote(x)
        url = 'https://habr.com/ru/search/?q=' + x 
        req = requests.get(url)
        time.sleep(0.3)
        soup = BeautifulSoup(req.text)
        articles = soup.find_all('article', class_='tm-articles-list__item')

        for article in articles:
            title = article.find('h2', class_='tm-article-snippet__title tm-article-snippet__title_h2').text
            link = 'https://habr.com' + article.find('h2', class_='tm-article-snippet__title tm-article-snippet__title_h2').find('a').get('href')
            date = article.find('span', class_='tm-article-snippet__datetime-published').text.strip()
            keyword = x
            row = {'keyword':keyword, 'date': date, 'title': title, 'link': link}
            habr_article = pd.concat([habr_article, pd.DataFrame([row])]) 
    
    return habr_article.reset_index(drop=True)

habr = get_habr_posts(['python', 'soft skills'])
habr
```

```python
habr.drop_duplicates(subset=['date', 'title', 'link'])
```

**Дополнительная часть**

Функция из обязательной части задания должна быть расширена следующим образом:

*   кроме списка ключевых слов для поиска необходимо объявить параметр с количеством страниц поисковой выдачи. Т.е. при передаче в функцию аргумента 4 необходимо получить материалы с первых 4 страниц результатов;
*   в датафрейме должны быть столбцы с полным текстом найденных материалов и количеством лайков:

<дата> - <заголовок> - <ссылка на материал> - <текст материала> - <количество лайков>


```python
search = ['python', 'data science', 'java']
```

```python
def get_habr_posts_full(query, pages):
    """Функция собирает с сайта все посты по ключевым словам и по количеству указанных страниц     
    """
    habr_article = pd.DataFrame()
    
    for x in query:             
        x = urllib.parse.quote(x)
        
        for p in range(1, pages + 1):            
            url = 'https://habr.com/ru/search/page'+ str(p) +'/?q=' + x
            req = requests.get(url)
            time.sleep(0.3)
            soup = BeautifulSoup(req.text)      
            articles = soup.find_all('article', class_='tm-articles-list__item')

            for article in articles:
                if article.find('h2', class_='tm-article-snippet__title tm-article-snippet__title_h2'):
                    title = article.find('h2', class_='tm-article-snippet__title tm-article-snippet__title_h2').text
                else:
                    pass
                if article.find('h2', class_='tm-article-snippet__title tm-article-snippet__title_h2'):
                    link = 'https://habr.com' + article.find('h2', class_='tm-article-snippet__title tm-article-snippet__title_h2').find('a').get('href')          
                else:
                    pass
                if  article.find('span', class_='tm-article-snippet__datetime-published'):   
                    date = article.find('span', class_='tm-article-snippet__datetime-published').text.strip()
                else:
                    pass
                if BeautifulSoup(requests.get(link).text).find('div', class_='tm-article-body'):
                    full_text = BeautifulSoup(requests.get(link).text).find('div', class_= 'tm-article-body').text.strip()
                else:
                    pass

                likes = article.find('span', class_='tm-votes-meter__value_rating').text
                keyword = x
                page_number = p
                row = {'keyword':keyword,'page_number':page_number, 'date': date, 'title': title, 'link': link, 'likes':likes,'full_text': full_text}
                habr_article = pd.concat([habr_article, pd.DataFrame([row])])

            p += 1

    return habr_article.reset_index(drop=True)

habr_full = get_habr_posts_full(search, 2)
habr_full
```
