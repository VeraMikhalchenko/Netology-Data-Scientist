```python
year = int(input("Введите год: "))

# программа, которая выведет, является ли данный год високосным или обычным.
if year % 4 == 0:
        print('Високосный год')
elif year % 4 != 0 or (year % 100 == 0 and year % 400 != 0):
        print('Невисокосный год')
else:
        print('Неверно введен год')
```

```python
# Программа, которая будет определять, является ли данный билет “счастливым”

number = 123456

number_str = str(number)
number_list = list(number_str)
sum_1 = int(number_list[0]) + int(number_list[1]) + int(number_list[2])
sum_2 = int(number_list[3]) + int(number_list[4]) + int(number_list[5])

if sum_1 == sum_2:
        print('счастливый билет!')
else:
        print('несчастливый билет')
```

```python
figure = str(input('Введите тип фигуры: круг, треугольник, прямоугольник:')).lower()

if figure == "круг":
        pi = 3.1415926535
        r = int(input('Введите радиус круга:'))
        S = pi * (r**2)
        print('Площадь круга =', round(S,2))
elif figure == "треугольник":
        A = int(input('Введите длину стороны А:'))
        B = int(input('Введите длину стороны B:'))
        C = int(input('Введите длину стороны C:'))
        p = (A + B + C)/2
        import math
        S = math.sqrt(p * (p-A) * (p-B) * (p-C))
        print('Площадь треугольника =', round(S,2))
elif figure == "прямоугольник":
        A = int(input('Введите длину стороны А:'))
        B = int(input('Введите длину стороны B:'))
        S = A * B
        print('Площадь прямоугольника =', S)
else:
        print('Неверный ввод')
```
