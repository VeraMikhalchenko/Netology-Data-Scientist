**Задание 1**

Скачайте с сайта grouplens.org...movielens/ датасет любого размера. Определите, какому фильму было выставлено больше всего оценок 5.0.

```python
import pandas as pd
```

```python
ratings = pd.read_csv ('ratings.csv')
ratings.head()
```

```python
ratings.info()
```

```python
best_movies = ratings[ratings['rating'] == 5.0] #создаю таблицу с лучшими фильмами
best_movies.head()
```

```python
best_movies['movieId'].value_counts() #считаю количество оценок для каждого из лучших фильмов
```

```python
best_movie_ID = best_movies['movieId'].value_counts().index[0] #нахожу ID фильма, который набрал наибольшее количество оценок 5.0
```

```python
print(f'фильму с ID {best_movie_ID} было выставлено больше всего оценок 5.0')
```

**Задание 2**

По данным файла power.csv посчитайте суммарное потребление стран Прибалтики (Латвия, Литва и Эстония) категорий 4, 12 и 21 за период с 2005 по 2010 года. Не учитывайте в расчетах отрицательные значения quantity.

```python
power = pd.read_csv ('power.csv')
power.head()
```

```python
power.info()
```

Фильтрую строки и проверяю

```python
Baltiya_power = power[(power['country'].isin(['Latvia','Lithuania','Estonia'])) 
                     & (power['year'].between(2005,2010))
                     & (power ['quantity'] >= 0)
                     & (power ['category'].isin([4, 12, 21]))]

Baltiya_power['country'].unique()
```

```python
Baltiya_power['year'].unique()
```

```python
Baltiya_power['quantity'].unique()
```

```python
Baltiya_power['category'].unique()
```

```python
Baltiya_power.head()
```

```python
sum_Baltiya = sum(Baltiya_power['quantity'])
print(f'суммарное потребление стран Прибалтики: {sum_Baltiya}')
```

```python
import io

import pytablereader as ptr
import pytablewriter as ptw

import retryrequests
```

```python
loader = ptr.TableUrlLoader(
        'https://ru.wikipedia.org/w/index.php?title=250_лучших_фильмов_по_версии_IMDb&stable=1',
        'html')

writer = ptw.TableWriterFactory.create_from_format_name('rst')
writer.stream = io.open('load_url_result.rst', 'w', encoding=loader.encoding)

for table_data in loader.load():
        writer.from_tabledata(table_data)
        best_films = table_data.as_dataframe()

best_films.head(250)
```
