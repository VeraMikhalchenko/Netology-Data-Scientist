Смоделировать игру против лотерейного автомата типа “777”. Игрок платит 1 руб., после чего выпадает случайное целое число, равномерно распределенное от 0 до 999. При некоторых значениях числа игрок получает выигрыш.

- 777: 200 руб.
- 999: 100 руб.
- 555: 50 руб.
- 333: 15 руб.
- 111: 10 руб.
- *77: 5 руб.
- **7: 3 руб.
- *00: 2 руб.
- **0: 1 руб.

Вопросы:

Выгодна ли игра игроку?
Сколько в среднем приобретает или теряет игрок за одну игру? \

```python
import numpy as np
```

Необходимо оговориться, что наше случайное число не совсем соответствует модели игрового автомата. Поскольку мы моделируем не 3 цифры как случайные величины от 0 до 9, а как одно число от 0 до 999. (было оговорено в вебинаре)

```python
trials = 10000000    # Количество попыток
successes = 0    # Количество успехов
waste = 0     # Считаем потери
winning = 0     # Считаем выйгрыш

for i in range(trials):
    random_score = np.random.randint(1, 1000)    # Случайное целое число от 0 до 999
    waste -= 1
    if random_score == 777:
        successes += 1
        winning += 200

    elif random_score == 999:
        successes += 1
        winning += 100

    elif random_score == 555:
        successes += 1
        winning += 50

    elif random_score == 333:
        successes += 1
        winning += 15
    
    elif random_score == 111:
        successes += 1
        winning += 10

    elif random_score % 100 == 77:
        successes += 1
        winning += 5
    
    elif random_score % 10 == 7:
        successes += 1
        winning += 3
    
    elif random_score % 100 == 0:
        successes += 1
        winning += 2
    
    elif random_score % 10 == 0:
        successes += 1
        winning += 1

balance = winning + waste # Выйгрыш/проигрыш
balance_per_game = balance / trials # Среднее за игру

p_monte_carlo = successes / trials    # Вероятность из моделирования                     

print(f'Успешные попытки: {successes},\nКол-во попыток: {trials},\nВероятность из моделирования: {p_monte_carlo}') 
print()

if balance > -(waste):
    print(f'Игра выгодна. Выйгрыш: {balance}р. В среднем игрок приоретает за игру: {balance_per_game}р.')
else:
    print(f'Игра не выгодна. Проигрыш: {balance}р. В среднем игрок теряет за игру: {balance_per_game}р.')
```

**Дополнительное задание повышенной сложности.**

Теоретически рассчитать средний выигрыш (проигрыш) и сравнить с результатами моделирования

Необходимо рассчитать математическое ожидание:

Вероятность выпадания чисел 777, 999, 555, 333, 111 = 1/кол-во вариантов = 0, 001 \
Вероятность выпадания чисел X77, X00 = 0,01\
Вероятность выпадания чисел XX7, XX0 = 0,09\
Вероятность выпадания остальных вариантов = 0,8

```python
n = 1000 # Кол-во вариантов
m_77 = 0 
m_7 = 0
m_00 = 0
m_0 = 0

for i in range(n):
    if i % 100 == 77:
        m_77 +=1 # Считаем, сколько раз может выпасть число *77
    elif i % 10 == 7:
        m_7 +=1 # Считаем, сколько раз может выпасть число **7
    elif i % 100 == 0:
        m_00 +=1 # Считаем, сколько раз может выпасть число *00
    elif i % 10 == 0:
        m_0 +=1 # Считаем, сколько раз может выпасть число **0

#Считаем вероятности для каждого из исходов:
p_77 = m_77 / n
p_7 = m_7 / n
p_00 = m_00 / n
p_0 = m_0 / n
p = (n - (m_77 + m_7 + m_00 + m_0)) / n

print(f'Вероятность выпадания *77 = {p_77}')
print(f'Вероятность выпадания **7 = {p_7}')
print(f'Вероятность выпадания *00 = {p_00}')
print(f'Вероятность выпадания **0 = {p_0}')
print(f'Вероятность остальных исходов = {p}')
```

Согласно формуле мат ожидания, умножим вероятности на стоимость выйгрыша/проигрыша и сложим:

```python
X_average = 0.001*((200 - 1) + (100 - 1) + (50 - 1) + (15 - 1) + (10 - 9))\
            + 0.01 * (5 - 1) + 0.09 * (3 - 1) + 0.01 * (2 - 1) + 0.09 * (1 - 1) + 0.8 * (0 - 1) #из каждого выйгрыша вычетаем потраченную сумму на 1 розыгрыш
X_average
```
